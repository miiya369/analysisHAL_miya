```python
import sys, os
import cmath
import numpy             as np
import numpy.linalg      as lin
import scipy             as sp
import scipy.integrate   as integ
import matplotlib.pyplot as plt

path_PythonCodeSet = "/Users/miiya/python_code-set"
sys.path.append(path_PythonCodeSet)

import common.statistics
import fitting.fitfunc_type
import fitting.io_params
import sch_diffeq.io_param_files
import sch_diffeq.solve_diffeq
import Tmatrix.io_Tmatrix
import Tmatrix.convert_mat
import Tmatrix.calc_phase_shift

hbar_c = 197.327053
```

```python
ibase = "/Users/miiya/Dropbox/programs/data/pacs-cs.LxLx_XxxN/c/ens1/fit"

ifname = "Pot.LcLc0_LcLc0.spin_0_.t13.bin_size45.fitparam"
```

```python
Nch = 2

FuncName, Params, mass = sch_diffeq.io_param_files.input_NxN_params(ibase+"/fitparam.summary")

fit_funcs = np.array([[fitting.fitfunc_type.set_fitfunc_from_fname(FuncName[ich][jch])
                       for jch in range(Nch)] for ich in range(Nch)])
```

```python
Emin   = 0.0
Edel   = 1
Emax   = 100.0

Rmax   = 5.0
Nplot  = 300

N_E   = int((Emax - Emin) / Edel)
Edata = np.array([Emin + i*Edel for i in range(N_E)])
Range = np.linspace(0, Rmax, Nplot)

Nconf  = len(Params[0,0,:,0])
Nparam = len(Params[0,0,0,:])

iconf = 3
tmpParams = np.array([[[Params[ich,jch,iconf,iparam]
                        for iparam in range(Nparam)] for jch in range(Nch)] for ich in range(Nch)])

Mvec = np.array([(mass[ich,0]*mass[ich,1]) / (mass[ich,0]+mass[ich,1]) for ich in range(Nch)])
Eth  = np.array([(mass[ich,0]+mass[ich,1]) - (mass[  0,0]+mass[  0,1]) for ich in range(Nch)])

InitPhi = np.zeros((2*Nch, Nch), dtype=float)
for ich in range(Nch):
    InitPhi[ich+Nch, ich] = 1.0

Evec = np.array([[Edata[iE] - Eth[ich] for ich in range(Nch)] for iE in range(N_E)])
Evec[Evec==0.0] = 1e-10

Phi = np.array([[integ.odeint(sch_diffeq.solve_diffeq.func_sch_ode, InitPhi[:,jch], Range, 
                              args=(fit_funcs, tmpParams, Evec[iE,:], Mvec)) 
                 for jch in range(Nch)] for iE in range(N_E)])
```

```python
rPhi = np.array([[[[Phi[iE,jch,iR,ich] for jch in range(Nch)] for ich in range(2*Nch)] 
                 for iR in range(Nplot)] for iE in range(N_E)])

Kvec = np.array([[cmath.sqrt(Evec[iE,ich]*Mvec[ich]*2.0) / hbar_c for ich in range(Nch)] for iE in range(N_E)])
sqKM = np.array([np.sqrt(Kvec[iE,:]*Mvec) for iE in range(N_E)])

KexM = np.array([[np.exp(-1j*(Kvec[iE,:]*Range[iR] - 0*np.pi/2.0)) for iR in range(Nplot)] for iE in range(N_E)])
KexP = np.array([[np.exp(+1j*(Kvec[iE,:]*Range[iR] - 0*np.pi/2.0)) for iR in range(Nplot)] for iE in range(N_E)])

JmK = np.array([[np.dot(np.diag(KexM[iE,iR,:]), 
                        rPhi[iE,iR,Nch:,:] + 1j * np.dot(np.diag(Kvec[iE,:]), 
                                                        rPhi[iE,iR,:Nch,:]))
                 for iR in range(Nplot)] for iE in range(N_E)])

JpK = np.array([[np.dot(np.diag(KexP[iE,iR,:]), 
                        rPhi[iE,iR,Nch:,:] - 1j * np.dot(np.diag(Kvec[iE,:]), 
                                                        rPhi[iE,iR,:Nch,:]))
                 for iR in range(Nplot)] for iE in range(N_E)])

Smat = np.array([np.dot(np.diag(1.0/sqKM[iE,:]), 
                        np.dot(JmK[iE,Nplot-1,:,:], 
                               np.dot(lin.inv(JpK[iE,Nplot-1,:,:]), 
                                      np.diag(sqKM[iE,:])))) 
                 for iE in range(N_E)])

Mat = np.array([Smat[iE,:,:] for iE in range(N_E)])
#Mat = np.array([0.5j*(np.identity(Nch)-Smat[iE,:,:]) for iE in range(N_E)])
#Mat = np.array([np.dot(Smat[iE,:,:], np.conjugate(Smat[iE,:,:].T)) for iE in range(N_E)])
```
