```python
import sys, os
import numpy             as np
import scipy             as sp
import scipy.optimize    as opt
import matplotlib.pyplot as plt
from   matplotlib        import patches

path_PythonCodeSet = "/Users/miiya/python_code-set"
sys.path.append(path_PythonCodeSet)

import common.io_data_bin
import common.statistics
import fitting.fitfunc_type
import fitting.io_params
import sch_GEM.solve_sch_GEM
import sch_GEM.set_mat_GEM
```

```python
ibase  = "/Users/miiya/Dropbox/programs/data/pacs-cs.LxLx_XxxN/c/ens1/fit"
ifname = "Pot.XccN0_XccN0.spin_0_.t15.bin_size57.fitparam"

fname, params = fitting.io_params.input_params(ibase+"/"+ifname)
```

```python
fig, axe = plt.subplots(nrows=2, ncols=2, figsize=(12,9))

#param   = np.mean(params, axis=0)
param   = params[2,:]
mu      = 1116.69
#mu      = 913.135
Nret    = 3
rmax    = 30.0
granges = np.linspace(0.5, 0.89, 50)
Nbase   = 50
Ranges  = [np.array([rmax*pow(granges[j], i) for i in range(Nbase)]) for j in range(len(granges))]
Eval    = np.array([sch_GEM.solve_sch_GEM.solve_sch_GEM(3, Ranges[i], param, mu, fname, 0, True)[0]
                  for i in range(len(granges))])
axe[0,0].plot(granges, Eval[:,0], 'b.', lw=3)
axe[0,0].set_xlim([0.55, 0.9])
axe[0,0].set_ylim([-1, 1])

grang = 0.85
Range = np.array([rmax*pow(grang, i) for i in range(Nbase)])
Eval  = np.array([sch_GEM.solve_sch_GEM.solve_sch_GEM(1, Range, params[i,:], mu, fname, 0, True)[0]
                  for i in range(len(params))])
axe[1,0].plot(np.arange(len(params)), Eval[:,0], 'b.', lw=3)
axe[1,0].set_xlim([-1, len(params)])
axe[1,0].set_ylim([-0.5, 0.5])
axe[1,0].set_ylim([-1, 1])

Eval_ave = np.mean(Eval[:,0])
Eval_err = np.std (Eval[:,0]) * np.sqrt(len(params)-1)
axe[1,0].add_patch(patches.Rectangle((0, Eval_ave-Eval_err), len(params)-1, 2*Eval_err, color='green', alpha=0.2))
axe[1,0].plot((0, len(params)-1), (Eval_ave, Eval_ave), lw=3, color='green')
print("Binding energy = %f (%f) [MeV]" % (Eval_ave, Eval_err))

rmax_plot = 10
r   = np.linspace(0, rmax_plot, 100)
ret = sch_GEM.solve_sch_GEM.solve_sch_GEM(1, Range, param, mu, fname, 0, False)
w   = np.array([np.sum(ret[1][:,0]*np.exp(-pow(r[ir]/Range,2))) for ir in range(len(r))])
axe[1,1].plot(r, fitting.fitfunc_type.set_fitfunc_from_fname(fname)(r,*param), lw=3)
axe[0,1].plot(r, w, lw=3)

for i in range(2):
    axe[i,1].set_xlim([0, rmax_plot])
    for j in range(2):
        axe[i,j].grid(which='major',color='gray',linestyle='--')
        axe[i,j].axhline(0.0, lw=2, color='black')
```
