```python
points, weights = np.polynomial.legendre.leggauss(100)

pcut = 1000 # [MeV]
pi_4 = np.pi / 4.0
pj   = pcut *                  np.tan(pi_4*(points+1.0))
sj   = pcut * pi_4 * weights / np.cos(pi_4*(points+1.0))**2
```

```python
fig, axe = plt.subplots(nrows=2, ncols=2, figsize=(12,9))

axe[0,0].plot(points, weights, 'b.', lw=2)
axe[0,1].plot(sj    ,      pj, 'b.', lw=2)
axe[1,0].plot(points,      pj, 'b.', lw=2)
axe[1,1].plot(points,      sj, 'b.', lw=2)



for i in range(2):
    for j in range(2):
        axe[i,j].grid(which='major',color='gray',linestyle='--')
        if (i+j != 0):
            axe[i,j].set_yscale("log")
            axe[i,j].axhline(pcut, lw=2, color='green')
axe[0,1].set_xscale("log")
axe[0,1].axvline(pcut, lw=2, color='green')
```
